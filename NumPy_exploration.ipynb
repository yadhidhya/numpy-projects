let's find out the time difference between lists and NumPy arrays.

```python
#lists
%%time
a1 = list(range(10000000))
a2 = list(range(10000000,20000000))
result=0
for x1,x2 in zip(a1,a2):
  result+=x1*x2
print(result)
```

```python
#numpy arrays
%%time
np_a1 = np.array(a1)
np_a2 = np.array(a2)
result=0
for x1,x2 in zip(np_a1,np_a2):
  result+=x1*x2
print(result)
```

**%%time** is an operator used to findout the time taken by the cpu to execute the code.

**What if the size of two matrices are different?**

Numpy array supports broadcasting, which allows arithmetic operations between two arrays having a different dimensions, but shapes should be compatible(i"ll explain it later)

```python
arr2=np.array([[[23,54,43],[34,45,53],[32,54,76]]])
arr3=np.array([10,10,10])
print(arr2)
print()
print('after addition')
print()
print(arr2+arr3)
```

here, you can observe that the values of 2nd row and 3rd row of arr2 were added by 10, which means rows of arr3 were duplicated to make their shape same

let's add the arrays with incompatible sizes

```python
arr2=np.array([[[23,54,43],[34,45,53],[32,54,76]]])
arr3=np.array([10,10])
print(arr2)
print()
print('after addition')
print()
print(arr2+arr3)
```

**here, you can observe that addition is not possible for these arrays because, their sizes are incompatible.**
so you got the error:


ValueError: operands could not be broadcast together with shapes (1,3,3) (2,)

from this we can conclude that, either the number of rows of  both the matrices or number of columns of both the matrices should be same to perform the broadcasting. other wise those matrices are considered as incompatible shaped matrices and leaves you an error message
